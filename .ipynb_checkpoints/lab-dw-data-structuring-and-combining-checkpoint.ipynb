## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd

def format_columns(dff):
    dff.columns=[col.replace(" ","_").lower() for col in dff.columns]
    return dff

def rename_col(dff):
    dff.rename(columns={"st":"state"}, inplace=True)
    return dff

def Gender(row):
    if row["gender"]=="Femal":
        return "F"
    elif row["gender"]=="Male":
        return "M"
    elif row["gender"]=="female":
        return "F"
    return row["gender"]
    
def replace_bachelor(dff):
    dff['education'] = dff['education'].replace({"Bachelors":"Bachelor"})
    return dff

def replace_character(dff):
    # Convertir les valeurs en chaînes de caractères
    dff['customer_lifetime_value'] = dff['customer_lifetime_value'].astype(str)
    
    # Remplacer le symbole % par rien
    dff['customer_lifetime_value'] = dff['customer_lifetime_value'].str.replace('%', '', regex=False)
    
    return dff

def convert_int(dff):
    dff['customer_lifetime_value'] = pd.to_numeric(dff['customer_lifetime_value'], errors='coerce').fillna(0).astype(int)
    return dff

def opencomplaints(dff):
    dff['number_of_open_complaints'] = dff['number_of_open_complaints'].apply(lambda x: x.split('/')[1] if isinstance(x, str) and '/' in x else x)
    dff['number_of_open_complaints'] = pd.to_numeric(dff['number_of_open_complaints'], errors='coerce').fillna(0).astype(int)
    return dff

def State(row):
    if row["state"] == "AZ":
        return "Arizona"
    elif row["state"] == "Cali":
        return "California"
    elif row["state"] == "WA":
        return "Washington"
    return row["state"]

def apply_state(df):
    df['state'] = df.apply(State, axis=1)
    return df

def fill_col(dff):
    num_col = dff.select_dtypes(include=['float64', 'int64']).columns
    cat_col = dff.select_dtypes(include=['object']).columns
    for col in num_col:
        dff[col] = dff[col].fillna(dff[col].median())
    for col in cat_col:
        dff[col] = dff[col].fillna(dff[col].mode()[0])
    return dff

# Function cleaning the 3 df
def clean_data(df):
    df = format_columns(df)
    df = rename_col(df)
    df['gender'] = df.apply(Gender, axis=1)
    df = replace_bachelor(df) 
    df = replace_character(df)
    df = convert_int(df)
    df = opencomplaints(df)
    df = apply_state(df)
    df = fill_col(df)
    return df




# 3 dataframes
df1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')

# Apply cleaning df
df1_cleaned = clean_data(df1)
df2_cleaned = clean_data(df2)
df3_cleaned = clean_data(df3)

# Concat 3 df
combined_df = pd.concat([df1_cleaned, df2_cleaned, df3_cleaned], ignore_index=True)

#Save data cleaned
combined_df.to_csv('combined_cleaned_data.csv', index=False)
```

```python
#1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. 
#Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail). 
#Round the total revenue to 2 decimal points. Analyze the resulting table to draw insights.

# Using pivot to show the total revenue for each sales channel
pivot_revenue = pd.pivot_table(df4, values='total_claim_amount', index='sales_channel', aggfunc='sum')

# Round to 2 decimal points
pivot_revenue = pivot_revenue.round(2)

# Display resultat
print(pivot_revenue)
```

```python
#2 Create a pivot table that shows the average customer lifetime value per gender and education level. 
#Analyze the resulting table to draw insights.

# Usinh pivot to find the average per gender and education leve of customer lifetime value
pivot_clv = pd.pivot_table(df4, values='customer_lifetime_value', index='gender', columns='education', aggfunc='mean')

# Round to 2 decimal points
pivot_clv = pivot_clv.round(2)

# Print the result
print(pivot_clv)
```

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Convert date to datetime type
df4['effective_to_date'] = pd.to_datetime(df4['effective_to_date'])

# Extrait month from
df4['month'] = df4['effective_to_date'].dt.month

# Check mont
print(df4[['effective_to_date', 'month']].head())
```

```python
# Using pivot for complaints per policy type & per month

df4['number_of_open_complaints'] = df4['number_of_open_complaints'].fillna(0)  # fill null values
complaints_pivot = pd.pivot_table(df4, values='number_of_open_complaints', index='policy_type', columns='month', aggfunc='sum')

# create a long format
complaints_long = complaints_pivot.reset_index().melt(id_vars='policy_type', var_name='month', value_name='number_of_open_complaints')

# Display table
print(complaints_long)
```
